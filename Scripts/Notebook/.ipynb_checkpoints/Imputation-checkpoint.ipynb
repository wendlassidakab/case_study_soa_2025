```python
# Sélection des variables et suppression des NA
tmp = df[["Distance.to.Nearest.City..km.", "Probability.of.Failure"]].dropna()

x = tmp["Distance.to.Nearest.City..km."]
y = tmp["Probability.of.Failure"]

# Régression linéaire
coef = np.polyfit(x, y, 1)
y_hat = np.polyval(coef, x)

# Scatter + droite de régression
plt.scatter(x, y, alpha=0.5)
plt.plot(x, y_hat, color = "red")

plt.xlabel("Distance to Nearest City (km)")
plt.ylabel("Probability of Failure")
plt.title("Probability of Failure vs Distance to Nearest City")
plt.show()
```

Plus on la distance augmente, plus la probabilité de défaillance diminue.

```python
# Variables à imputer
vars_to_impute = [
    "Height..m.",
    "Length..km.",
    "Volume..m3.",
    "Drainage..km2.",
    "Surface..km2.",
    "Distance.to.Nearest.City..km."
]

# Variables à exclure de l'imputation 
vars_excluded = [
    "Probability of Failure",
    "Loss given failure – prop (Qm)",
    "Loss given failure – liab (Qm)",
    "Loss given failure – BI (Qm)",
    "Height_clean",
    "Length_clean",
    "Length_missing",
    "Volume_clean",
    "Volume_missing",
    "Volume_log",
    "n_modifications",
    "last_modification_year"
]
```

```python
for v in vars_to_impute:
    df[v + "_missing"] = df[v].isna().astype(int)
```

```python
imputation_features = [
    c for c in df.columns
    if c not in vars_excluded
]
```

```python
df_imp_base = df[imputation_features].select_dtypes(include=["number"])

all_na_cols = df_imp_base.columns[df_imp_base.isna().all()]

df_imp_base2 = df_imp_base.drop(columns=all_na_cols)
```

```python
imputer = IterativeImputer(
    estimator=RandomForestRegressor(
        n_estimators=300,
        max_depth=10,
        random_state=42,
        n_jobs=-1
    ),
    max_iter=10,
    random_state=42
)

df_imp_array = imputer.fit_transform(df_imp_base2)

df_imp = pd.DataFrame(
    df_imp_array,
    columns=df_imp_base2.columns,
    index=df.index
)
```

```python
df_imputed = df.copy()

for v in vars_to_impute:
    df_imputed[v] = df_imp[v]

df_imputed = df_imputed.drop(columns=all_na_cols)
```

```python
# Proportion de NA après imputation
df_imputed[vars_to_impute].isna().mean()
```

```python
# Comparaison avant / après
df[vars_to_impute].describe()
```

## Estimation des pertes 

```python
features = [
    "Regulated.Dam", "Primary.Purpose", "Primary.Type", "Height..m.", "Length..km.", "Year.Completed", "Surface..km2.",
    "Drainage..km2.", "Spillway", "Distance.to.Nearest.City..km.", "Hazard", "Assessment", "n_modifications"
]
```

```python
target = "Loss.given.failure...BI..Qm."

df_train = df_imputed[df_imputed[target].notna()].copy()
df_pred  = df_imputed[df_imputed[target].isna()].copy()

X_train = df_train[features]
y_train = df_train[target]

X_pred = df_pred[features]
```

```python
num_cols = X_train.select_dtypes(include=["number"]).columns
cat_cols = X_train.select_dtypes(exclude=["number"]).columns

num_cols = X_train.select_dtypes(include=["number"]).columns
cat_cols = X_train.select_dtypes(exclude=["number"]).columns

preprocess = ColumnTransformer(
    transformers=[
        (
            "cat",
            OneHotEncoder(handle_unknown="ignore"),
            cat_cols
        )
    ],
    remainder="passthrough"
)
```

```python
models = {
    "Linear": LinearRegression(),

    "RF": RandomForestRegressor(
        n_estimators=300,
        max_depth=10,
        random_state=42,
        n_jobs=-1
    ),

    "GB": GradientBoostingRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    ),

    "XGB": XGBRegressor(
        n_estimators=400,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="reg:squarederror",
        random_state=42,
        n_jobs=-1
    )
}
```

```python
results = {}

for name, model in models.items():
    pipe = Pipeline([
        ("prep", preprocess),
        ("model", model)
    ])
    
    mae = -cross_val_score(
        pipe,
        X_train,
        y_train,
        cv=5,
        scoring="neg_mean_absolute_error"
    ).mean()
    
    results[name] = mae

results
```

```python
best_model_name = min(results, key=results.get)
best_model = models[best_model_name]

final_pipe = Pipeline([
    ("prep", preprocess),
    ("model", best_model)
])

final_pipe.fit(X_train, y_train)
```

```python
df_imputed.loc[
    df_imputed[target].isna(),
    target
] = final_pipe.predict(X_pred)
```
