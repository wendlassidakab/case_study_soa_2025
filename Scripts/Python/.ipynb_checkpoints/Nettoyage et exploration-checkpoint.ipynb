```python
# Fonction pour obtenir les statistiques descriptives des variables des données
def summarize_df(Dataframe):

    summary = pd.DataFrame()

    # Type de données 
    summary["Data Type"] = Dataframe.dtypes

    # Nombre de valeur uniques
    summary["Unique values"] = Dataframe.nunique()

    # Les valeurs uniques 
    summary["Unique"] = Dataframe.apply(lambda x: x.unique().tolist())

    # Maximum
    summary["Max"] = Dataframe.apply(lambda x: x.max() if pd.api.types.is_numeric_dtype(x) else "-")

    # Minimum 
    summary["Min"] = Dataframe.apply(lambda x: x.min() if pd.api.types.is_numeric_dtype(x) else "-")

    # Moyenne, Médiane, Mode
    summary["Mean"] = Dataframe.apply(lambda x: round(x.mean(), 4) if pd.api.types.is_numeric_dtype(x) else "-")
    summary["Median"] = Dataframe.apply(lambda x: x.median() if pd.api.types.is_numeric_dtype(x) else "-")
    summary["Mode"] = Dataframe.apply(lambda x: x.mode().iloc[0] if pd.api.types.is_numeric_dtype(x) else "-")

    # Mesures de dispersion
    summary["Range"] = Dataframe.apply(lambda x: x.max() - x.min() if pd.api.types.is_numeric_dtype(x) else "-")
    summary["Variance"] = Dataframe.apply(lambda x: x.var() if pd.api.types.is_numeric_dtype(x) else "-")
    summary["Range"] = Dataframe.apply(lambda x: x.std() if pd.api.types.is_numeric_dtype(x) else "-")

    # Mesures de la forme de la distribution 
    summary["Skewness"] = Dataframe.apply(lambda x: round(x.skew(), 4) if pd.api.types.is_numeric_dtype(x) else "-")
    summary["Kurtosis"] = Dataframe.apply(lambda x: round(x.kurt(), 4) if pd.api.types.is_numeric_dtype(x) else "-")

    return summary
```

```python
summarize_df(df)
```

## Hauteur des barrages 

```python
(df_2["Height (m)"] == 0).mean() * 100
```

La proportion de barrages présentant une hauteur égale à zéro est extrêmement faible (environ 0,06 %). Cette valeur marginale suggère que ces observations correspondent très probablement à des informations manquantes ou à des cas particuliers d’ouvrages de très faible hauteur, plutôt qu’à des structures réellement dépourvues de hauteur. Compte tenu de leur faible fréquence, ces valeurs n’ont pas d’impact significatif sur l’analyse globale, mais sont néanmoins traitées avec prudence dans les analyses quantitatives continues afin d’éviter tout biais.

```python
df_2["Height_clean"] = df_2["Height (m)"].replace(0, np.nan)
df_2["Height_clean"].describe()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))

sns.histplot(
    df_2["Height_clean"].dropna(),
    bins=30
)

plt.xlabel("Hauteur du barrage (m)")
plt.ylabel("Nombre de barrages")
plt.title("Distribution de la hauteur des barrages en terre")
plt.tight_layout()
plt.show()
```

```python
plt.boxplot(df_2["Height_clean"].dropna(), vert=False)
plt.xlabel("Hauteur du barrage (m)")
plt.title("Boxplot de la hauteur des barrages")
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))

sns.boxplot(
    data=df_2,
    x="Region",
    y="Height_clean",
    order=["Flumevale", "Lyndrassia", "Navaldia"],
    showfliers=True
)

plt.ylabel("Hauteur du barrage (m)")
plt.xlabel("Région")
plt.title("Distribution de la hauteur des barrages par région")
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.displot(
    data=df_2,
    x="Height_clean",
    col="Region",
    col_order=["Flumevale", "Lyndrassia", "Navaldia"],
    bins=30,
    kde=False
)

g.set_axis_labels("Hauteur du barrage (m)", "Nombre de barrages")
g.set_titles("Région : {col_name}")
plt.tight_layout()
plt.show()
```

L’analyse descriptive montre que les barrages en terre sont majoritairement de faible à moyenne hauteur. La hauteur moyenne est d’environ 10,8 m, tandis que la médiane (~9,0 m) confirme que plus de la moitié des ouvrages restent relativement bas. La distribution est fortement asymétrique à droite, avec une grande majorité de barrages concentrés sous 15 m (Q1 ~ 6,6 m ; Q3 ~ 11,8 m) et la présence de quelques ouvrages très élevés pouvant dépasser 200 m, ce qui explique l’écart important entre la moyenne et la médiane ainsi que la forte dispersion observée.

Les histogrammes et boxplots par région mettent en évidence des différences régionales : Flumevale et Navaldia présentent une dispersion plus large et davantage de valeurs extrêmes, tandis que Lyndrassia affiche des hauteurs plus homogènes et généralement plus faibles. Ces résultats suggèrent des profils de risque hétérogènes selon la région, où la hauteur des barrages constitue un facteur clé de sévérité potentielle en cas de défaillance.

```python
df_2["Length (km)"].describe()
```

La longueur du barrage présente une distribution fortement asymétrique, avec une médiane de 0,21 km et une majorité d’ouvrages de faible extension.
```python
df_2.groupby("Region")["Length_clean"].describe()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df_2,
    x="Region",
    y="Length_clean",
    showfliers=True
)

plt.ylabel("Longueur du barrage (km)")
plt.xlabel("Région")
plt.title("Distribution de la longueur des barrages en terre par région")
plt.tight_layout()
plt.show()
```

```python
missing_by_region = (
    df_2.groupby("Region")["Length_missing"]
    .mean()
    .reset_index()
)

missing_by_region["Length_missing"] *= 100
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(
    data=missing_by_region,
    x="Region",
    y="Length_missing"
)

plt.ylabel("Proportion de longueurs non renseignées (%)")
plt.xlabel("Région")
plt.title("Part des longueurs de barrage non renseignées par région")
plt.tight_layout()
plt.show()
```

L’analyse de la longueur des barrages met en évidence des différences structurelles marquées entre les régions. Lyndrassia se caractérise par des ouvrages plus courts et homogènes, tandis que Navaldia présente des barrages nettement plus longs et une forte dispersion, incluant des infrastructures exceptionnellement étendues. Par ailleurs, la proportion de longueurs non renseignées varie fortement selon les régions, atteignant près de 30 % à Lyndrassia, ce qui indique une qualité de données hétérogène. Afin de limiter les biais et de préserver l’information, les longueurs nulles ont été traitées comme manquantes et un indicateur spécifique a été introduit pour la modélisation.

## Volume 

```python
df_2["Volume (m3)"].describe()
```

```python
df["Volume (m3)"].isna().mean() * 100
```

```python
(df["Volume (m3)"] == 0).mean() * 100
```

```python
df_2["Volume_clean"] = df_2["Volume (m3)"].replace(0, np.nan) # remplacer les 0 par nan

df_2["Volume_missing"] = df_2["Volume_clean"].isna().astype(int) # Indicateur de valeur manquante 

df_2["Volume_log"] = np.log1p(df_2["Volume_clean"]) # Valeurs logarithmique étant donné les grandes valeurs prises par la variable 
```

```python
df_2.groupby("Region")["Volume_log"].describe()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df_2,
    x="Region",
    y="Volume_log",
    showfliers=True
)

plt.ylabel("log(Volume du barrage + 1)")
plt.xlabel("Région")
plt.title("Distribution du volume des barrages (échelle logarithmique) par région")
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(
    df_2,
    col="Region",
    col_wrap=3,
    height=4,
    sharex=True,
    sharey=True
)

g.map_dataframe(
    sns.histplot,
    x="Volume_log",
    bins=30,
    kde=True
)

g.set_axis_labels("log(Volume du barrage + 1)", "Nombre de barrages")
g.fig.suptitle("Distribution du volume des barrages par région (échelle logarithmique)", y=1.05)
plt.show()
```

```python
# Proportion des volumes manquants 
volume_missing_by_region = df_2.groupby("Region")["Volume_missing"].mean().reset_index()

volume_missing_by_region["Volume_missing"] *= 100

plt.figure(figsize=(7, 5))
sns.barplot(
    data=volume_missing_by_region,
    x="Region",
    y="Volume_missing"
)

plt.ylabel("Proportion de volumes non renseignés (%)")
plt.xlabel("Région")
plt.title("Part des volumes de barrage non renseignés par région")
plt.tight_layout()
plt.show()
```

Le volume du barrage est une variable centrale dans cette étude de cas, car il constitue un indicateur direct de la capacité de stockage d’eau et, par conséquent, du niveau de dommages potentiels en cas de défaillance. Dans une perspective de gestion des risques et d’analyse actuarielle, cette variable est étroitement liée à la notion de loss given failure, puisque des volumes plus élevés impliquent généralement des conséquences économiques, humaines et environnementales plus importantes.

L’exploration des données met toutefois en évidence plusieurs enjeux de qualité. Près de la moitié des observations présentent un volume non renseigné, et une proportion non négligeable de barrages affiche un volume égal à zéro. Ces valeurs nulles peuvent correspondre soit à de très petits ouvrages, soit à une information manquante codée par défaut. Dans ce contexte, la suppression des observations aurait conduit à une perte substantielle d’information et à un biais potentiel dans l’analyse. Le choix retenu consiste donc à conserver l’ensemble des barrages et à introduire un indicateur spécifique de données manquantes, afin de préserver la structure du portefeuille.

La distribution du volume brut est fortement asymétrique, avec une concentration des valeurs à des niveaux relativement faibles et quelques barrages de très grande capacité qui tirent la moyenne vers le haut. Cette asymétrie justifie l’utilisation d’une transformation logarithmique du volume, qui permet de réduire l’influence des valeurs extrêmes, de stabiliser la variance et de rendre les distributions plus comparables entre régions. 

L’analyse par région révèle des différences marquées. Navaldia se distingue par des volumes en moyenne plus élevés et une dispersion importante, suggérant la présence de barrages de grande envergure. Flumevale présente une variabilité notable, avec un mélange de petits et moyens ouvrages, tandis que Lyndrassia affiche des volumes globalement plus faibles et plus homogènes. Ces contrastes régionaux indiquent que l’exposition au risque n’est pas uniformément répartie et doivent être pris en compte dans toute analyse comparative ou décisionnelle.

Enfin, la répartition des volumes non renseignés varie fortement selon les régions, avec une proportion particulièrement élevée à Navaldia et Lyndrassia. Cette hétérogénéité suggère des différences dans les pratiques de collecte ou dans la nature des infrastructures recensées. D’un point de vue méthodologique, le maintien des observations accompagné d’un indicateur de données manquantes constitue une approche robuste et conforme aux bonnes pratiques, permettant d’exploiter au mieux l’information disponible tout en limitant les biais lors de la modélisation.
